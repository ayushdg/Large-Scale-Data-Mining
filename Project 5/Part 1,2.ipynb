```python
import json
import gzip
import tarfile
import numpy as np
import datetime, time
import pytz
import pickle
from matplotlib import pyplot as plt
import statsmodels.api as sm
from sklearn import preprocessing
import math
```

```python
load_data = True

if not load_data:
    
    tar = tarfile.open("data/tweet_data.tar.gz", "r:gz")
    
    avgFollowers = {}
    tweetTimes = {}
    avgRetweets = {}

    pst_tz = pytz.timezone('US/Pacific')
    
    for member in tar.getmembers():
        f = tar.extractfile(member)
        followerData = []
        timeData = []
        retweetData = []
        for line in f.readlines():
            entry = json.loads(line)
            #Follower Data
            followerData.append(entry['author']['followers'])
            #Tweet time Data
            citation = entry['citation_date']
            timeData.append(datetime.datetime.fromtimestamp(citation, pst_tz))
            #Retweet Data
            retweetData.append(entry['metrics']['citations']['total'])
        
        avgFollowers[member.name] = (sum(followerData) / len (followerData))
        avgRetweets[member.name] =  (sum(retweetData) / len(retweetData))
        tweetTimes[member.name] = timeData
        
        f.close()
    
    tar.close()
    
    with open('data/train/part1/part_1_1_data.pkl' , 'wb') as file:
        pickle.dump([avgFollowers, tweetTimes, avgRetweets] , file)
    
else:
    with open('data/train/part1/part_1_1_data.pkl' , 'rb') as file:
        [avgFollowers, tweetTimes, avgRetweets] = pickle.load(file)
    
```

```python
def getHourlyTweets(tweetTimes):
    hourlyTweetCount = {}
    
    for i in range(14,32):
        for j in range(0,24):
            hourlyTweetCount[str(i) + ' '+ str(j)] = 0
    
    for i in range(1,8):
        for j in range(0,24):
            hourlyTweetCount[str(i) + ' '+ str(j)] = 0
    
    for tweet in tweetTimes:
        day = tweet.day
        hour = tweet.hour
        currentCount = hourlyTweetCount[str(day) + ' ' + str(hour)]
        hourlyTweetCount[str(day) + ' ' + str(hour)] = currentCount + 1
        
    return hourlyTweetCount
```

```python
load_data = True

if not load_data:
    
    tar = tarfile.open("data/tweet_data.tar.gz", "r:gz")
    
    featureData = {}

    pst_tz = pytz.timezone('US/Pacific')
    
    for member in tar.getmembers():
        f = tar.extractfile(member)
        featureValues = []
        for line in f.readlines():
            entry = json.loads(line)
            #Follower Data
            authorFollowers = entry['author']['followers']
            #Tweet time Data
            citation = entry['citation_date']
            tweetTime = datetime.datetime.fromtimestamp(citation, pst_tz)
            #Retweet Data
            retweetCount = entry['metrics']['citations']['total']
            
            featureValues.append([authorFollowers , tweetTime, retweetCount])
        
        featureData[member.name] = featureValues
        
        f.close()
    
    tar.close()
    
    with open('data/train/part1/part_1_2_data.pkl' , 'wb') as file:
        pickle.dump(featureData , file)
    
else:
    with open('data/train/part1/part_1_2_data.pkl' , 'rb') as file:
        featureData = pickle.load(file)
    
```

```python
def getHourlyFeatures(featureData):
    hourlyFeatures = {}
    
    for i in range(14,32):
        for j in range(0,24):
            hourlyFeatures[str(i) + ' '+ str(j)] = {'tweetCount' : 0, 'retweetCount' : 0, 'followerCount' : 0,
                                                    'maxFollowers' : 0, 'timeOfDay' : j}
    
    for i in range(1,8):
        for j in range(0,24):
            hourlyFeatures[str(i) + ' '+ str(j)] = {'tweetCount' : 0, 'retweetCount' : 0, 'followerCount' : 0,
                                                    'maxFollowers' : 0, 'timeOfDay' : j}
    
    for feature in featureData:
        day = feature[1].day
        hour = feature[1].hour
        followers = feature[0]
        retweets = feature[2]
        
        tweetCount = hourlyFeatures[str(day) + ' ' + str(hour)]['tweetCount']
        retweetCount = hourlyFeatures[str(day) + ' ' + str(hour)]['retweetCount']
        followerCount = hourlyFeatures[str(day) + ' ' + str(hour)]['followerCount']
        currentMax = hourlyFeatures[str(day) + ' ' + str(hour)]['maxFollowers']
        
        hourlyFeatures[str(day) + ' ' + str(hour)]['tweetCount'] = tweetCount + 1
        hourlyFeatures[str(day) + ' ' + str(hour)]['retweetCount'] = retweetCount + retweets
        hourlyFeatures[str(day) + ' ' + str(hour)]['followerCount'] = followerCount + followers
        hourlyFeatures[str(day) + ' ' + str(hour)]['maxFollowers'] = max(currentMax , followers)
        
    return hourlyFeatures
```

```python
def generateFeatureVectors(hourlyFeatures):
    X = []
    y = []
    for i in range(15,32):
        for j in range(0,24):
            #For the first hour of the day we take the last hour of the previous day
            if j == 0:
                jj = 23
                ii = i -1
            else:
                jj = j - 1
                ii = i
                
            y.append([hourlyFeatures[str(i) + ' '+ str(j)]['tweetCount']])
            
            X.append([hourlyFeatures[str(ii) + ' '+ str(jj)]['tweetCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['retweetCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['followerCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['maxFollowers'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['timeOfDay']])
            
    for i in range(1,8):
        for j in range(0,24):
            #For the first hour of the day we take the last hour of the previous day
            if j == 0:
                jj = 23
                ii = i - 1
                if ii == 0:
                    ii = 31
            else:
                jj = j - 1
                ii = i
                
            y.append([hourlyFeatures[str(i) + ' '+ str(j)]['tweetCount']])
            
            X.append([hourlyFeatures[str(ii) + ' '+ str(jj)]['tweetCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['retweetCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['followerCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['maxFollowers'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['timeOfDay']])
            
    #for the first day we dont want hour 1 prediction
    for j in range(1,24):

        i = 14
        ii = 14
        jj = j - 1
                
        y.append([hourlyFeatures[str(i) + ' '+ str(j)]['tweetCount']])
            
        X.append([hourlyFeatures[str(ii) + ' '+ str(jj)]['tweetCount'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['retweetCount'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['followerCount'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['maxFollowers'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['timeOfDay']])
            
        
    return [np.array(X), np.array(y)]
    
```

## 1.2 Regression Results

```python
feature_list = ['tweetCount','retweetCount', 'followerCount',
                 'maxFollowers', 'timeOfDay']
for key in featureData.keys():
    [X, y] = generateFeatureVectors(getHourlyFeatures(featureData[key]))
    X = preprocessing.scale(X)
    X = sm.add_constant(X, prepend=False)
    mod = sm.OLS(y, X)
    res = mod.fit()
    y_d = res.predict()
    y_res = np.sum(np.square(y - y_d)) / len(y)
    rmse = np.sqrt(y_res)
    print('\n The regression results for the tweets with the hashtag %s are '% key[7:-4])
    print(res.summary())
    best_features = res.pvalues.argsort()[:4]
    best_features = best_features.tolist()
    if 5 in best_features:
        best_features.remove(5)
    print ("\nBest features selected are:\n")
    for index in range(0,3):
            print('%s \n' % feature_list[best_features[index]])
    print('\nRMSE : %f' % rmse)
```

```python
load_data = True

if not load_data:
    
    tar = tarfile.open("data/tweet_data.tar.gz", "r:gz")
    
    featureData = {}

    pst_tz = pytz.timezone('US/Pacific')
    
    for member in tar.getmembers():
        f = tar.extractfile(member)
        featureValues = []
        for line in f.readlines():
            entry = json.loads(line)
            #Follower Data
            authorFollowers = entry['author']['followers']
            #Tweet time Data
            citation = entry['citation_date']
            tweetTime = datetime.datetime.fromtimestamp(citation, pst_tz)
            #Retweet Data
            retweetCount = entry['metrics']['citations']['total']
            #Impressions
            impressionCount = entry['metrics']['impressions']
            #RankingScore
            rankingScore = entry['metrics']['ranking_score']
            #Number of #Tags in the tweet
            hashTags = entry['title'].count('#')
            #Favourite count
            favCount = entry['tweet']['favorite_count']
            #Numbet of mentions of the tweet
            userMentions = len(entry['tweet']['entities']['user_mentions'])
            #Number of URL in the tweet
            urls = len(entry['tweet']['entities']['urls'])
            #Author ID
            authId = entry['author']['nick']
            #Long Tweet boolean greater than 90 char
            longTweet = len(entry['title']) > 90
            
            featureValues.append([authorFollowers , tweetTime, retweetCount,impressionCount,
                                  rankingScore, hashTags, favCount, userMentions,urls,authId,
                                  longTweet])
        
        featureData[member.name] = featureValues
        
        f.close()
    
    tar.close()
    
    with open('data/train/part1/part_1_3_data.pkl' , 'wb') as file:
        pickle.dump(featureData , file)
    
else:
    with open('data/train/part1/part_1_3_data.pkl' , 'rb') as file:
        featureData = pickle.load(file)
    
```

```python
def getNewHourlyFeatures(featureData):
    hourlyFeatures = {}
    
    for i in range(14,32):
        for j in range(0,24):
            hourlyFeatures[str(i) + ' '+ str(j)] = {'tweetCount' : 0, 'retweetCount' : 0, 'followerCount' : 0,
                                                    'maxFollowers' : 0, 'timeOfDay' : [0] * 24, 'impressionCount':0,
                                                    'rankingScore':0, 'hashTag':0, 'favCount':0, 'mentions':0, 'urls' : 0,
                                                    'author' : [], 'longTweets' : 0}
            hourlyFeatures[str(i) + ' '+ str(j)]['timeOfDay'][j] = 1
    
    for i in range(1,8):
        for j in range(0,24):
            hourlyFeatures[str(i) + ' '+ str(j)] = {'tweetCount' : 0, 'retweetCount' : 0, 'followerCount' : 0,
                                                    'maxFollowers' : 0, 'timeOfDay' : [0] * 24, 'impressionCount':0,
                                                    'rankingScore':0, 'hashTag':0, 'favCount':0, 'mentions':0, 'urls' : 0,
                                                    'author' : [], 'longTweets' : 0}
            hourlyFeatures[str(i) + ' '+ str(j)]['timeOfDay'][j] = 1
    
    for feature in featureData:
        day = feature[1].day
        hour = feature[1].hour
        followers = feature[0]
        retweets = feature[2]
        impressionCount = feature[3]
        rankingScore = feature[4]
        hashTags = feature[5]
        favCount = feature[6]
        mentions = feature[7]
        urls = feature[8]
        author = feature[9]
        longTweet = feature[10]
        
        tweetCount = hourlyFeatures[str(day) + ' ' + str(hour)]['tweetCount']
        retweetCount = hourlyFeatures[str(day) + ' ' + str(hour)]['retweetCount']
        followerCount = hourlyFeatures[str(day) + ' ' + str(hour)]['followerCount']
        currentMax = hourlyFeatures[str(day) + ' ' + str(hour)]['maxFollowers']
        currentImp = hourlyFeatures[str(day) + ' ' + str(hour)]['impressionCount']
        currentRank = hourlyFeatures[str(day) + ' ' + str(hour)]['rankingScore']
        currentTags = hourlyFeatures[str(day) + ' ' + str(hour)]['hashTag']
        currentFavs = hourlyFeatures[str(day) + ' ' + str(hour)]['favCount']
        currentMen = hourlyFeatures[str(day) + ' ' + str(hour)]['mentions']
        currentUrl = hourlyFeatures[str(day) + ' ' + str(hour)]['urls']
        currentLong = hourlyFeatures[str(day) + ' ' + str(hour)]['longTweets']
        
        hourlyFeatures[str(day) + ' ' + str(hour)]['tweetCount'] = tweetCount + 1
        hourlyFeatures[str(day) + ' ' + str(hour)]['retweetCount'] = retweetCount + retweets
        hourlyFeatures[str(day) + ' ' + str(hour)]['followerCount'] = followerCount + followers
        hourlyFeatures[str(day) + ' ' + str(hour)]['maxFollowers'] = max(currentMax , followers)
        hourlyFeatures[str(day) + ' ' + str(hour)]['impressionCount'] = currentImp + impressionCount
        hourlyFeatures[str(day) + ' ' + str(hour)]['rankingScore'] = currentRank + rankingScore
        hourlyFeatures[str(day) + ' ' + str(hour)]['hashTag'] = currentTags + hashTags
        hourlyFeatures[str(day) + ' ' + str(hour)]['favCount'] = currentFavs + favCount
        hourlyFeatures[str(day) + ' ' + str(hour)]['mentions'] = currentMen + mentions
        hourlyFeatures[str(day) + ' ' + str(hour)]['urls'] = currentUrl + urls
        hourlyFeatures[str(day) + ' ' + str(hour)]['author'].append(author)
        hourlyFeatures[str(day) + ' ' + str(hour)]['longTweets'] = currentLong + (1 if longTweet else 0)
        
    return hourlyFeatures
```

```python
def generateNewFeatureVectors(hourlyFeatures):
    X = []
    y = []
    for i in range(15,32):
        for j in range(0,24):
            #For the first hour of the day we take the last hour of the previous day
            if j == 0:
                jj = 23
                ii = i -1
            else:
                jj = j - 1
                ii = i
            fVec = []
            y.append([hourlyFeatures[str(i) + ' '+ str(j)]['tweetCount']])
            
            fVec.extend([hourlyFeatures[str(ii) + ' '+ str(jj)]['tweetCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['retweetCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['followerCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['maxFollowers']])
            fVec.extend(hourlyFeatures[str(ii) + ' '+ str(jj)]['timeOfDay'])
            fVec.extend([hourlyFeatures[str(ii) + ' '+ str(jj)]['impressionCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['rankingScore'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['hashTag'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['favCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['mentions'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['urls'],
                      len(set(hourlyFeatures[str(ii) + ' '+ str(jj)]['author'])),
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['longTweets']])
            X.append(fVec)
            
    for i in range(1,8):
        for j in range(0,24):
            #For the first hour of the day we take the last hour of the previous day
            if j == 0:
                jj = 23
                ii = i - 1
                if ii == 0:
                    ii = 31
            else:
                jj = j - 1
                ii = i
                
            fVec = []
            y.append([hourlyFeatures[str(i) + ' '+ str(j)]['tweetCount']])
            
            fVec.extend([hourlyFeatures[str(ii) + ' '+ str(jj)]['tweetCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['retweetCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['followerCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['maxFollowers']])
            fVec.extend(hourlyFeatures[str(ii) + ' '+ str(jj)]['timeOfDay'])
            fVec.extend([hourlyFeatures[str(ii) + ' '+ str(jj)]['impressionCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['rankingScore'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['hashTag'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['favCount'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['mentions'],
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['urls'],
                      len(set(hourlyFeatures[str(ii) + ' '+ str(jj)]['author'])),
                      hourlyFeatures[str(ii) + ' '+ str(jj)]['longTweets']])
            X.append(fVec)
            
    #for the first day we dont want hour 1 prediction
    for j in range(1,24):

        i = 14
        ii = 14
        jj = j - 1
                
        fVec = []
        y.append([hourlyFeatures[str(i) + ' '+ str(j)]['tweetCount']])

        fVec.extend([hourlyFeatures[str(ii) + ' '+ str(jj)]['tweetCount'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['retweetCount'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['followerCount'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['maxFollowers']])
        fVec.extend(hourlyFeatures[str(ii) + ' '+ str(jj)]['timeOfDay'])
        fVec.extend([hourlyFeatures[str(ii) + ' '+ str(jj)]['impressionCount'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['rankingScore'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['hashTag'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['favCount'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['mentions'],
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['urls'],
                  len(set(hourlyFeatures[str(ii) + ' '+ str(jj)]['author'])),
                  hourlyFeatures[str(ii) + ' '+ str(jj)]['longTweets']])
        X.append(fVec)
            
        
    return [np.array(X), np.array(y)]
    
```

```python
feature_list = ['tweetCount','retweetCount', 'followerCount',
                 'maxFollowers']
feature_list.extend(['timeOfDay'] * 24)
feature_list.extend(['impressionCount' , 'rankingScore','hashTags','favourite count','mentions','urls','authors','longTweets'])
for key in featureData.keys():
    [X, y] = generateNewFeatureVectors(getNewHourlyFeatures(featureData[key]))
    X = preprocessing.scale(X)
    X = sm.add_constant(X, prepend=False)
    mod = sm.OLS(y, X)
    res = mod.fit()
    y_d = res.predict()
    y_res = np.sum(np.square(y - y_d)) / len(y)
    rmse = np.sqrt(y_res)
    print('\n The regression results for the tweets with the hashtag %s are '% key[7:-4])
    print(res.summary())
    best_features = res.pvalues.argsort()[:30]
    best_features = best_features.tolist()
    if 36 in best_features:
        best_features.remove(36)
    for timeOfDay in range(4,28):
        if timeOfDay in best_features:
            best_features.remove(timeOfDay)
    print ("\nBest features selected are:\n")
    #print(best_features)
    for index in range(0,3):
            print('%s \n' % feature_list[best_features[index]])
            plt.title("Scatter plot for {0} against {1}".format(key[7:-4], feature_list[best_features[index]]))
            plt.xlabel("Number of tweets/hour")
            plt.ylabel(feature_list[best_features[index]])
            plt.scatter(y, X[:, best_features[index]])
            plt.plot([min(y), max(y)], [min(X[:, best_features[index]]), max(X[:, best_features[index]])], 
                     color='green', marker='o', linestyle='dashed')
            plt.show()
    print('\nRMSE : %f' % rmse)
```
